# Programação Computacional para Engenharia

## Lista de Exercícios 01 - Estrutura Sequencial

### Objetivos
O objetivo desta lista é permitir que os alunos pratiquem o conteúdo apresentado na aula de conceitos básicos de programação, utilizando a linguagem Python. Os principais conceitos apresentados tratam de leitura e impressão de valores, assim como alguns operadores matemáticos.

### Enunciados

1.   Faça um programa que imprime na tela a frase "*Programar em Python é divertido!*".



2.   Faça um programa que leia o seu primeiro nome e imprima-o na tela.



3.   Faça um programa que peça dois números inteiros e imprima a soma desses dois números.




4.  Dois carros (X e Y) partem na mesma direção. O carro X sai com uma velocidade constante de 60 km/h e o carro Y sai com uma velocidade constante de 90 km/h. Em uma hora (60 minutos) o carro Y pode distanciar-se 30 quilômetros do carro X. Leia a distância (em km) e calcule quanto tempo leva (em minutos) para o carro Y percorrer essa distância em relação ao outro carro.

Exemplo de Entrada: 30 / Exemplo de Saída: 60 minutos

5. Implemente um programa que leia 3 valores referentes à quantidade de dias, horas e minutos. Calcule o tempo total em segundos.

6. Faça o processo inverso ao da questão 5 e leia um número inteiro que represente a quantidade de segundos e transforme esse valor em horas, minutos e segundos.

Exemplo de Entrada: 

> 556

Exemplo de Saída: 

> hora(s): 0<br>
> minuto(s): 9<br>
> segundos: 16
